# - Loading and Cleaning the data

<hr style="border:2px solid black"> </hr>

- Imports and loading data

```python
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.formula.api as smf
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_graphviz
from sklearn.metrics import confusion_matrix
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
vgsales = pd.read_csv("vgsales.csv")
vgsales
```

- Checking information for all the columns

```python
vgsales.describe()
```

```python
vgsales.describe(include = "O")
```

- Removing NaN values

```python
nan_filter = vgsales.isna().any(axis=1)
vgsales[nan_filter]
```

```python
missing_year = vgsales["Year"].isna()
vgsales_year = vgsales[~ missing_year]
missing_publisher = vgsales_year["Publisher"].isna()
vgsales2 = vgsales_year[~ missing_publisher]
vgsales2
```

- Removing rows with year values greater than 2016

```python
vgsales2.value_counts("Year")
```

```python
index_year = vgsales2[ vgsales2['Year'] > 2016 ].index
index_year
```

```python
vgsales2.drop(index_year, inplace = True) 
vgsales2
```

- Removing any duplicates (Need for Speed: Most Wanted is a case where it's a different game with the same name that came out in a different year)

```python
dup_filter = vgsales2.duplicated(subset=['Name','Platform'])
vgsales2[dup_filter]
```

```python
dup_filter = vgsales2.duplicated(subset=['Name','Platform','Year'])
vgsales2[dup_filter]
```

```python
vgsales2 = vgsales2.drop_duplicates(subset=['Name','Platform','Year'])
vgsales2
```

# - Model 1: Decision Tree

<hr style="border:2px solid black"> </hr>

```python
vgsales2
```

Will be using the Genre column. I will create dummy variables and split the data into testing and training data.

- Create dummy variables

```python
vgsales3 = pd.get_dummies(vgsales2, columns = ["Genre"], drop_first = True)
vgsales3
```

```python
x = vgsales3.drop(columns = ["NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales","Name","Platform","Publisher"])
x.head()
```

```python
y = vgsales3["Global_Sales"]
y.head()
```

- Split data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

- Fit a model max depth 3

```python
tree = DecisionTreeRegressor(max_depth = 3)
tree.fit(x_train, y_train)
```

```python
preds_test = tree.predict(x_test)
```

```python
mean_squared_error(preds_test, y_test)
```

```python
preds_train = tree.predict(x_train)
```

```python
mean_squared_error(preds_train,y_train)
```

- fit a model max depth 10

```python
tree10 = DecisionTreeRegressor(max_depth = 10)
tree10.fit(x_train, y_train)
```

```python
preds_test10 = tree10.predict(x_test)
```

```python
mean_squared_error(preds_test10, y_test)
```

```python
preds_train10 = tree10.predict(x_train)
```

```python
mean_squared_error(preds_train10,y_train)
```

- fit a model max depth 20

```python
tree20 = DecisionTreeRegressor(max_depth = 20)
tree20.fit(x_train, y_train)
```

```python
preds_test20 = tree20.predict(x_test)
```

```python
mean_squared_error(preds_test20, y_test)
```

```python
preds_train20 = tree20.predict(x_train)
```

```python
mean_squared_error(preds_train20,y_train)
```

# - Sales Distribution By Region and Genre Using Heatmap

<hr style="border:2px solid black"> </hr>

```python
vgsales4 = vgsales2.groupby(vgsales2["Genre"]).sum()
vgsales4 = vgsales4.drop(columns = ["Rank", "Year", "Global_Sales"])
vgsales4
```

```python
vgsales4_transposed = vgsales4.transpose()
vgsales4_transposed
```

```python
plt.figure(figsize=(20, 8))
plt.yticks(
    verticalalignment = 'center',
)
ax = sns.heatmap(vgsales4_transposed, cmap = "YlGnBu", linewidths = 1)
plt.ylabel("Regions")
plt.title("Sales Distribution By Region and Genre")
plt.show()
```
